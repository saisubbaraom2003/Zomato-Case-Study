```python
df["Cuisines"]=df["Cuisines"].fillna("North Indian")#filling null values in Cusines column with high frequency value
```

```python
#importing Country Names with codes  and merging country names with zomato dataset by linking the tables with country codes
df_country=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Data_Sets/Country-Code.xlsx")
```

```python
df_country
```

```python
df_final=pd.merge(df,df_country,on="Country Code",how="outer")
```

```python
#find duplicates and remove it and find null entries for the merged data
df_final
```

```python
#checking the duplicates:
df_final.duplicated().sum()
```

```python
#checking null values:
df_final.isnull().sum()
```

### Top 3 countries that uses zomato:

```python
country_name=df_final["Country"].value_counts().index
```

```python
country_name
```

```python
df_final["Country"].value_counts()
```

```python
country_val=df_final["Country"].value_counts().values
country_val
```

```python
#visualising top3 countries:
col=["Green","Orange","Red"]
plt.pie(country_val[:3],labels=country_name[:3],colors=col,autopct="%1.2f%%")
plt.legend()
```

Observation: Zomato maximum records or transactions are from India after that USA and then UK.

### Analysis of zomato ratings:

```python
ratings=df_final.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

```python
plt.rcParams["figure.figsize"]=(12,6)   #plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,palette="pastel")
```

```python
#create a bar graph on the basis of rating colors:
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=ratings,palette=["Black","red","orange","yellow","green","pink"])
```

Observation:

1. Not rated count is very high
2. avrage ratings counts are falling between 2.5 to 3.4

### Name the countries that has given 0 ratings:

```python
df_final[df_final["Rating color"]=="White"].groupby("Country").size()
```

Maximum number of 0 ratings are from Indian customers.

# Top 10 cuisines and ploting the graph for it.

```python
df_cuisines=df_final["Cuisines"].value_counts().head(10).reset_index().rename(columns={0:"Rating count"})
```

```python
df_cuisines
```

```python
plt.figure(figsize=(13,6))
ax=sns.barplot(x="Cuisines",y="count",data=df_cuisines,palette="pastel")
for i in ax.containers:
  ax.bar_label(i,fmt="%d")
plt.xticks(fontsize=10, rotation=90)
plt.show()
```
